```python
#Figure 3G
file = pd.read_excel('Paper1_figure_3G.xls', sheet_name = "Sheet1", skiprows = 2)
df = pd.DataFrame(file)

values = []
col_names = []
xs = []
for i, col in enumerate(df.columns):
    values.append(df[col].values)
    col_names.append(col)
    xs.append(np.random.normal(i + 1, 0.04, df[col].values.shape[0]))

#remove the nan values
for i in range (int(len(values))):
    for k, j in enumerate(values[i]):
        if j != j :
            values[i][k] = 0    
print(np.std(values[0]))
print(np.std(values[1]))
print(np.std(values[2]))
file.plot.box()
plt.title("H2kb")
plt.ylabel('gMFI (x1000)')

palette = ['black', 'blue', 'red']
for x, val, c in zip(xs, values, palette):
    plt.scatter(x, val, alpha=1.0, color=c)
plt.yticks([0, 5, 10, 15])

plt.plot(3, 8, marker = "*", markerfacecolor = "k", markeredgecolor = "k")
plt.plot(2.9, 8, marker = "*", markerfacecolor = "k", markeredgecolor = "k")
plt.show()
```

```python
#Figure 5C 
#Unpaired t test
newFile4 = pd.read_excel('Paper1_figure_5C.xls', skiprows = 2)
df8 = pd.DataFrame(newFile4)
l = []
List = []
for _, col in enumerate(df8.columns):
    l.append(df8[col].values)
    
for i in l[0]:
    List.append([i])
    
for j in l[1:]:
    for i, val in enumerate(j):
        List[i].append(val)

for i, el in enumerate(List):
    List[i] = [el[0], el[1:5], el[5:9]]

#remove the nan values
List[3][1][-1] = 0
List[3][2][-1] = 0
print(List[1])
```

```python
A = []
Y1 = []
Y2 = []
Y1_error = []
Y2_error = []
for i in List:
    A.append(i[0])
    Y1.append(np.mean(i[1]))
    Y2.append(np.mean(i[2]))
    
X_axis = np.arange(len(A))
# plt.rcParams["figure.figsize"] = [7, 40]
plt.bar(X_axis - 0.2, Y1, 0.4, label = "21%O2", color = "black")
plt.bar(X_axis + 0.2, Y2, 0.4, label = "1.5%O2", color = "red")
plt.xticks(X_axis, A)
plt.yticks([0, 2, 4, 6, 8, 10, 30, 50])
plt.legend()
plt.plot(3.2, 10, marker = "*", markerfacecolor = "k", markeredgecolor = "k")
plt.plot(1.1, 50, marker = "*", markerfacecolor = "k", markeredgecolor = "k")
plt.plot(1.2, 50, marker = "*", markerfacecolor = "k", markeredgecolor = "k")
plt.plot(1.3, 50, marker = "*", markerfacecolor = "k", markeredgecolor = "k")
plt.plot(1.4, 50, marker = "*", markerfacecolor = "k", markeredgecolor = "k")
plt.show()
```

```python
#Unpaired-t test

#For BMDC
group1 = List[0][1]
group2 = List[0][2]
t, pVal= stats.ttest_ind(group1, group2)
if pVal < 0.05:
    print("BMDC is significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)
else:
    print("BMDC is not significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)

#For BMDC+SL8
group1 = List[1][1]
group2 = List[1][2]
t, pVal= stats.ttest_ind(group1, group2)
if pVal < 0.05:
    print("BMDC+SL8 is significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)
else:
    print("BMDC+SL8 is not significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)

#For TAM
group1 = List[2][1]
group2 = List[2][2]
t, pVal= stats.ttest_ind(group1, group2)
if pVal < 0.05:
    print("TAM is significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)
else:
    print("TAM is not significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)

#For 2xTAM
group1 = List[3][1]
group2 = List[3][2]
t, pVal= stats.ttest_ind(group1, group2)
if pVal < 0.05:
    print("2xTAM is significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)
else:
    print("2xTAM is not significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)

#For TAM
group1 = List[4][1]
group2 = List[4][2]
t, pVal= stats.ttest_ind(group1, group2)
if pVal < 0.05:
    print("TAM is significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)
else:
    print("TAM is not significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)

#For 2xTAM
group1 = List[5][1]
group2 = List[5][2]
t, pVal= stats.ttest_ind(group1, group2)
if pVal < 0.05:
    print("2xTAM is significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)
else:
    print("2xTAM is not significantly different between 21% O2 and 1.5% O2")
    print("pVal = ", pVal)
```
